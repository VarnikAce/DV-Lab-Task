```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Load Dataset ---
# Replace with your dataset path
df = pd.read_csv("/content/Employee (1).csv")

# --- 2. Identify categorical and continuous attributes ---
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
continuous_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("Categorical Attributes:", categorical_cols.tolist())
print("Continuous Attributes:", continuous_cols.tolist())

# Choose one categorical and one continuous column for demonstration
cat_col = categorical_cols[0] if len(categorical_cols) > 0 else None
cont_col = continuous_cols[0] if len(continuous_cols) > 0 else None

# --- 3. Univariate Analysis for Categorical Data ---
if cat_col:
    # Bar Chart
    plt.figure(figsize=(6,4))
    df[cat_col].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f"Bar Chart of {cat_col}")
    plt.xlabel(cat_col)
    plt.ylabel("Count")
    plt.show()

    # Pie Chart
    plt.figure(figsize=(6,6))
    df[cat_col].value_counts().plot(kind='pie', autopct='%1.1f%%',
    colors=sns.color_palette("pastel"))
    plt.title(f"Pie Chart of {cat_col}")
    plt.ylabel("") # Remove y-label for cleaner look
    plt.show()

# --- 4. Univariate Analysis for Continuous Data ---
if cont_col:
    # Histogram
    plt.figure(figsize=(6,4))
    sns.histplot(df[cont_col], bins=20, kde=False, color='orange')
    plt.title(f"Histogram of {cont_col}")
    plt.xlabel(cont_col)
    plt.ylabel("Frequency")
    plt.show()

    # Density Plot
    plt.figure(figsize=(6,4))
    sns.kdeplot(df[cont_col], shade=True, color='green')
    plt.title(f"Density Plot of {cont_col}")
    plt.xlabel(cont_col)
    plt.ylabel("Density")
    plt.show()

    # Rug Plot
    plt.figure(figsize=(6,4))
    sns.rugplot(df[cont_col], color='red')
    plt.title(f"Rug Plot of {cont_col}")
    plt.xlabel(cont_col)
    plt.show()
```
